```python
bond.rename(mapper = {"GoldenEye" : "Golden Eye"}, axis = 0, inplace=True)
bond.rename(columns={ "Year" : "Release Date"}, inplace=True)
bond.head(3)
```

```python
bond.drop("Casino Royale")
bond.pop("Actor")  #returns the series
del bond["Director"]
```

```python
bond.sample(3, axis = "columns")
```

```python
bond.nsmallest(3, columns="Box Office")
bond.nlargest(3, columns="Box Office")
bond["Box Office"].nlargest()
```

```python
mask = bond["Actor"] == "Sean Connery"
bond[mask]
bond.where(mask)  #takes a boolean value in the arguments
```

```python
bond.columns = [column_name.replace(" ", "_") for column_name in bond.columns]
bond.head(3)
```

```python
bond.query('Actor == "Sean Connery"')
bond.query('Director == "Terence Young"')
bond.query("Actor == 'Roger Moore' or Director == 'John Glenn'")
bond.query("Box_Office > 600")
bond.query("Actor in ['Timothy Dalton', 'George Lazenby']")
```

```python
def convert(number):
    return str(number) + " MILLIONS!"

bond["Box Office"].apply(convert)
bond["Budget"].apply(convert)


columns = ["Box Office", "Budget", "Bond Actor Salary"]
for cols in columns:
    bond[cols].apply(convert)
```

```python
def good_movie(row):

    actor = row[1]
    budget = row[4]

    if actor == 'Pierce Brosnan':
        return "The Best"
    elif actor == 'Roger Moore' and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue!"
    
bond.apply(good_movie, axis = "columns")
```
